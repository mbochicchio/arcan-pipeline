```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the filtered CSV file containing the project data
filtered_file = 'filtered_projects.csv'

# Load the data into a pandas DataFrame
df = pd.read_csv(filtered_file)

# Define the columns corresponding to the metrics
# The column continuous_integration and management have all values equal to 0 --> not included in the metrics_columns
metrics_columns = ['community', 'documentation', 
                   'history', 'license', 'unit_test', 'pull', 'releases']
```

```python
# Function to plot the distribution (histogram) and boxplot for each metric
def plot_metric_distribution_and_boxplot(df, metrics_columns):
    # Set Seaborn style for better-looking plots
    sns.set(style="whitegrid")
    
    for metric in metrics_columns:
        # Create a figure with two subplots (one for histogram, one for boxplot)
        fig, axes = plt.subplots(1, 2, figsize=(14, 6))
        
        # Plot the histogram
        axes[0].hist(df[metric], bins=20, edgecolor='black')
        axes[0].set_title(f'Distribution of {metric.capitalize()} (Histogram)')
        axes[0].set_xlabel(f'{metric.capitalize()}')
        axes[0].set_ylabel('Frequency')
        axes[0].grid(True)

        # Plot the boxplot
        sns.boxplot(ax=axes[1], x=df[metric], color="skyblue")
        axes[1].set_title(f'Distribution of {metric.capitalize()} (Boxplot)')
        axes[1].set_xlabel(f'{metric.capitalize()}')
        
        # Show the plots
        plt.tight_layout()
        plt.show()
```

```python
# Call the function to plot both the distributions and boxplots of all metrics
plot_metric_distribution_and_boxplot(df, metrics_columns)
```

```python
# Block 1: Distribution of Programming Languages
def plot_language_distribution(df):
    plt.figure(figsize=(10, 6))
    language_counts = df['language'].value_counts()
    sns.barplot(x=language_counts.index, y=language_counts.values, palette='viridis')
    plt.title('Distribution of Programming Languages')
    plt.xlabel('Programming Language')
    plt.ylabel('Number of Repositories')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Plot the distribution of programming languages
plot_language_distribution(df)
```

```python
# Block 2: Distribution of Size
def plot_size_distribution(df):
    plt.figure(figsize=(10, 6))
    sns.histplot(df['size'], bins=20, kde=True, color='orange')
    plt.title('Distribution of Project Size')
    plt.xlabel('Project Size')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Plot the distribution of project size
plot_size_distribution(df)
```

```python
# Calculate the number and percentage of projects with 0 value for each metric
def calculate_zero_projects(df, metrics_columns):
    total_projects = len(df)
    print(f"Total number of projects: {total_projects}\n")
    
    for metric in metrics_columns:
        zero_count = (df[metric] == 0).sum()  # Count how many projects have 0 value for the metric
        zero_percentage = (zero_count / total_projects) * 100  # Calculate the percentage
        print(f"Metric: {metric.capitalize()}")
        print(f"Projects with 0 value: {zero_count}")
        print(f"Percentage of projects with 0 value: {zero_percentage:.2f}%\n")

# Call the function to display the counts and percentages for each metric
calculate_zero_projects(df, metrics_columns)
```
